```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['sv','sev','c','cl','vdc','eng','it','evpr'])
y=np.array([79,56,98,64,46,39,88,93])

plt.plot(x,y,color='blue',marker='*')
plt.show() #line graph
plt.plot(x, y)
plt.grid()
plt.show()#line graph with grid
plt.xlabel("subjects")
plt.ylabel("marks")
plt.bar(x,y,color='hotpink')
plt.show()#bar ghaph with labels
plt.scatter(x,y)
plt.show()#points graph
a=['sv','sev','c','cl','vdc','eng','it','evpr']
plt.pie(y,labels=a)
plt.show()#pie chart
plt.hist(y)
plt.show()#histogram
plt.plot(x,y,linestyle='dashed',color='blue')
plt.show()#dashed line graph
```
